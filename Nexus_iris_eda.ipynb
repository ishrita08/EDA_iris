```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define column names
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]

# Load the Iris dataset
iris = pd.read_csv('/content/Iris.csv',skiprows=1, header=None, names=column_names)

```

```python
iris.head()
```

```python
iris.describe()
```

```python
plt.figure(figsize=(12, 6))
for i, feature in enumerate(column_names[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x="class", y=feature, data=iris)
    plt.title(f'Distribution of {feature} by Class')
plt.tight_layout()
plt.show()
```

```python
# Drop the 'class' column before calculating the correlation matrix
numeric_columns = iris.drop('class', axis=1)
correlation_matrix = numeric_columns.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Display information about the numeric columns
print("Info about numeric columns:")
print(numeric_columns.info())

# Display the first few rows of the numeric columns
print("First few rows of numeric columns:")
print(numeric_columns.head())
```

```python
# Drop the 'class' column before calculating the correlation matrix
numeric_columns = iris.drop('class', axis=1)
correlation_matrix = numeric_columns.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
